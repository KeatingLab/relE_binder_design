```python
data_path = '/data1/groups/keatinglab/MST_workspace/PixelDB/clusters'
path = '/home/gridsan/sswanson/keatinglab_shared/pixelDB_data/210926_manuallySelectedPixelDBComplexes_list.txt'
names = []
with open(path,'r') as file:
    for line in file:
        name = line.rstrip().split('/')[-1]
        names.append(name)
print(len(names),names[0:5])
file_list_path = 'pixelDB30_list.txt'
with open(file_list_path,'w') as file:
    for i,row in df[df['name'].isin(names)].iterrows():
        pdb_path = data_path+'/'+str(row.cluster_number)+'/'+str(row['name'])
        file.write(pdb_path+'\n')
```

## Filter out PixelDB30 (and homologs) from TERMinator training set 

```python
pixelDB30 = set([name.split('_')[0].lower() for name in names])
clusters = set([int(name.split('_')[3]) for name in names])
print(clusters)
pixelDB30hom_filt_df = df[df['cluster_number'].isin(clusters)]
pixelDB30hom_filt_df
```

```python
pixelDB30hom = set([x.split('_')[0].lower() for x in pixelDB30hom_filt_df.name])
pixelDB30hom
```

```python
in_path = '/data1/groups/keatinglab/fosterb/data/multichain_features/train.in'
out_path = '/data1/groups/keatinglab/fosterb/data/multichain_features/train_filt.in'
in_training_set = []
with open(in_path,'r') as in_file, open(out_path,'w') as out_file:
    for line in in_file:
        pdb_id = line.rstrip()
        if pdb_id not in pixelDB30hom:
            out_file.write(line)
        else:
            in_training_set.append(pdb_id)
print(len(in_training_set))
```

```python
# how many of the exact PixelDB30 structures were in the training set?
PixelDB30_intrainingset = [x for x in in_training_set if x in pixelDB30]
print(len(PixelDB30_intrainingset),PixelDB30_intrainingset)
```
